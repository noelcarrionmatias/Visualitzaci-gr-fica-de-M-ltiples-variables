```python
#Ejercicio 2 - Fes les tasques de preprocessing i adequació del dataset que disposem en el repositori de GitHub PRE-PROCESSING-DATA, movies.dat    amb l'objectiu de preparar-lo i treballar-lo com a dataframe per a extreure'n informació.
    #Definimos el nombre de las columnas (se ha visto que no tenian)
colname = ['Id','Film','Gen']
    #Abrimos el archivo, definiendo los separadores, el tipo de codigo, si tiene o no encabezado y el nombre de las columnas
df = pd.read_csv('C:/Users/ncarrion/Dropbox/Python/data_science_python/Modulo 2/T2/movies.txt', sep="::", encoding='utf-8',header=None, names=colname)
    #Se visualizan las 5 primeras filas
df.head()
```

```python
    #Ahora se obtiene una columna única para el año de la pelicula (Film)
df["Año"] = df['Film'].apply(lambda x: x[-5:-1])
df["Gen"].str.split('|', expand=True)
df.head()
```

```python
    #Separamos los diferentes generos de la película por separado
df['Gen1'], df['Gen2'], df['Gen3'], df['Gen4'], df['Gen5'] = df['Gen'].str.split('|',4).str
print(df)
```
